# CONVERTED 0 AND 1 VALUES OF SENIOR CITIZEN TO YES AND NO TO MAKE IT EASIER TO UNDERSTAND

```python
a["SeniorCitizen"].replace(to_replace=0,value="No",inplace=True)
a["SeniorCitizen"].replace(to_replace=1,value="Yes",inplace=True)
```

```python
r=sns.countplot(x="Churn",data=a)
r.bar_label(r.containers[0])
plt.title("Count of Customer by Churn")
plt.show()
```

```python

plt.figure(figsize=(3,4))
gb=a.groupby("Churn").agg({"Churn":"count"})
l=["No","Yes"]
plt.pie(gb["Churn"],labels=l,autopct="%1.2f%%")
plt.title("Percentage of Customer by Churn")
plt.show()
```

```python
sns.countplot(x="gender",data=a,hue="Churn")
plt.title("Count of Churn by gender")
plt.show()
```

```python
s=sns.countplot(x="SeniorCitizen",data=a)
s.bar_label(s.containers[0])
plt.title("Count of senior citizen")
plt.show()
```
