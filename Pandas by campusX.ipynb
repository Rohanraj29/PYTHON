# SERIES FROM INTEGER

```python
import pandas as pd
import numpy as np
age=[19,18,16,15]
a=pd.Series(age)
a
```

# creating a custom index and also provides a name to our series.

```python
import numpy as np
import pandas as pd
marks=[92,94,98,96,89]
subject=["English","Maths","Hindi","Science","Computer"]
b=pd.Series(marks,index=subject,name="Rahul Marks in 10th")
b
```
