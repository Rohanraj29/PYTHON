```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
super=pd.read_csv(r"C:\Users\rohan\Downloads\SuperStoreUS.csv")
super.head(2)
```

# first step is to check null values

```python
super.isnull().sum()
```

```python
mean=super["Product Base Margin"].mean()
```

```python
super["Product Base Margin"].fillna(mean,inplace=True)
```

# replace the value

```python
super["Order Priority"]=super["Order Priority"].replace('Critical ' ,"Critical")
```

```python
super["Order Priority"].unique()
```

```python
super["Order Priority"].value_counts()
```

# count of order priority. it shows that low priority of order occured most times in the dataset.

```python
sns.countplot(x="Order Priority",data=super)
plt.title("Count of order priority")
plt.show()
```

```python
super["Ship Mode"].value_counts().plot(kind="pie")
```

# count by ship mode. it shows that regular air is in the most demand.

```python
label=super["Ship Mode"].value_counts().index
value=super["Ship Mode"].value_counts().values
```

```python
plt.pie(value,labels=label,startangle=60,autopct="%0.2f%%")
plt.legend(loc=4)
plt.show()
```

# count of product category by ship mode.

```python
sns.countplot(x="Ship Mode",data=super,hue="Product Category")
plt.show()
```

# count of ship mode by order priority.

```python
sns.countplot(x="Ship Mode",data=super,hue="Order Priority")
```

```python
super["Customer Segment"].value_counts()
```

# count of customer segment. it shows that most of the people are from corporate segment.

```python
sns.countplot(x="Customer Segment",data=super)
plt.title("Count by Customer Segment")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
super=pd.read_csv(r"C:\Users\rohan\Downloads\SuperStoreUS.csv")
super.head(2)
```

# total profit by product category. it shows that office supplies product receives more profit.

```python
sns.barplot(x="Product Category",y="Profit",data=super,estimator="sum")
plt.show()
```

#  total sales by ship mode. it shows that regular air has highest sales.

```python
sns.barplot(x="Ship Mode",y="Sales",data=super,estimator="sum")
plt.show()
```

# top 5 states.

```python
super["State or Province"].value_counts()[:5].plot(kind="pie",autopct="%0.2f%%")
plt.show()
```

# total sales by product category. Technology has highest number of sales.

```python
sns.barplot(x="Product Category",y="Sales",data=super,estimator="sum")
plt.show()
```
