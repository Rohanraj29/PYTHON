# dataframe using list

```python
import numpy as np
import pandas as pd
details=[["Rahul",19,"Muzaffarpur"],["Rohan",18,"Muzaffarpur"],["Sakshi",16,"Muzaffarpur"],["Pihu",12,"Muzaffarpur"]]
a=pd.DataFrame(details,columns=["name","age","city"])
a
```

# dataframe using dictionary

```python
detail={
    "name":["Rahul","Rohan","Sakshi","Pihu"],
    "age":[19,18,16,15],
    "city":["Muzaffarpur","Muzaffarpur","Muzaffarpur","Muzaffarpur"]
}
r=pd.DataFrame(detail)
r.set_index("name",inplace=True)
r
```
