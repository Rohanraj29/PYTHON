```python
import matplotlib.pyplot as plt
a=["C","C++","Java","Python"]
b=[10,20,30,40]
plt.bar(a,b,width=0.4,color="orange",edgecolor="yellow",linewidth=4,alpha=0.9,label="Programming language",align="edge")
plt.legend()
plt.show()
```

```python
# 2. SCATTER PLOT
import matplotlib.pyplot as plt
a=[2,4,6,8]
b=[10,20,30,40]
plt.scatter(a,b,color="orange",s=80,edgecolor="red",linewidth=4,marker="o",alpha=0.4)
plt.show()
```

```python
# 3. HISTOGRAM PLOT
import matplotlib.pyplot as plt
import numpy as np
import random
mi=np.random.randint(10,40,(5))
csk=np.random.randint(0,25,(5))
bi=[10,15,25,35]
plt.hist([mi,csk],color=["green","yellow"],range=(20,35),bins=bi,rwidth=10,bottom=10,cumulative=-1,orientation="horizontal",label=["Mumbai indians","Chennai super kings"])
plt.legend()
```

```python
# 3. PIE CHART
import matplotlib.pyplot as plt
a=[120,400,800,720]
b=["C","C++","Java","Python"]
ex=[0.1,0.0,0.0,0.0]
plt.pie(a,labels=b,colors=["yellow","green","pink","orange"],explode=ex,shadow=True,radius=1.4,labeldistance=1.8,startangle=180,rotatelabels=360,autopct="%0.1f%%")
plt.legend()
plt.show()
```

```python
# 4. STEM PLOT
import matplotlib.pyplot as plt
a=[2,3,5,7]
b=[10,20,30,40]
plt.stem(a,b,linefmt=":",basefmt="green",markerfmt="pink",bottom=20,orientation="horizontal",label="Data Visualization")
plt.legend()
plt.show()
```

```python
# 5. AREA PLOT
import matplotlib.pyplot as plt
a=[2,4,6,8]
b=[10,15,20,25]
c=[20,35,45,55]
plt.stackplot(a,b,c,colors=["green","yellow"],labels=["Data1","Data2"])
plt.legend()
```

```python
# 6. STEP PLOT
import matplotlib.pyplot as plt
a=[2,4,6,8]
b=[10,20,30,40]
plt.step(a,b,ms=18,mfc="green",label="Data",marker="o",color="yellow")
plt.legend()
plt.show()
```

```python
# 7. SUBPLOT
import matplotlib.pyplot as plt
a=[2,4,6,8]
b=[4,8,12,16]
plt.subplot(2,2,1)
plt.plot(a,b)
c=[10,20,30,40]
d=[20,40,60,80]
plt.subplot(2,2,2)
plt.plot(c,d)
e=[50,70,90,110]
f=["Python","Java","C","C++"]
plt.subplot(2,2,3)
plt.pie(e,labels=f)
g=[19,18,16,15]
h=["Rahul","Rohan","Sakshi","Pihu"]
plt.subplot(2,2,4)
plt.pie(g,labels=h)
plt.show()
```

```python
# 8. LINE PLOT
import matplotlib.pyplot as plt
a=[2,4,6,8]
b=[10,15,20,25]
plt.plot(a,b,color="green",linestyle=":",marker="o",linewidth=4)
plt.show()
```

```python
# 9. XTICK AND XLIM
import matplotlib.pyplot as plt
a=[2,4,6,8]
b=[8,16,24,32]
plt.plot(a,b)
plt.xticks(a)
plt.yticks(b)
plt.xlim(0,12)
plt.ylim(12,35)
plt.show()
```
